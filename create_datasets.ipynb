```python
# Setup some contants

from pathlib import Path

INPUT_DATA = Path('data/raw_data.jsonl.xz')
TEST_SET = Path('data/test.csv')
TRAIN_SET = Path('data/train.csv')
```

```python
# Load data/sample.jsonl into pandas dataframe and sort by close_time in ascending order

import lzma
import pandas as pd

with lzma.open(INPUT_DATA, 'r') as file:
    df = pd.read_json(file, lines=True)
df = df.sort_values('close_time')
df.head()
```

```python
# Split the close_time_human column from datatime into year, month, day, hour

df['year'] = pd.DatetimeIndex(df['close_time_human']).year
df['month'] = pd.DatetimeIndex(df['close_time_human']).month
df['day'] = pd.DatetimeIndex(df['close_time_human']).day
df['hour'] = pd.DatetimeIndex(df['close_time_human']).hour
df.head()
```

```python
# Remove close_time_human, ledger_index, ledger_hash, close_flags, close_time, close_time_resolution, parent_close_time, parent_hash, transaction_hash, total_coins columns

df = df.drop([
    'close_time_human',
    'ledger_hash',
    'ledger_index',
    'close_flags',
    'close_time',
    'close_time_resolution',
    'parent_close_time',
    'parent_hash',
    'transaction_hash',
    'total_coins'
], axis=1, errors='ignore')
df.head()
```

```python
# Sum tx_count of rows that have the same year, month, day, and hour

df = df.groupby(['year', 'month', 'day', 'hour'], as_index=False).sum()
df.head()
```

```python
# Add columns that take the sines of month, day, and hour

import numpy as np

df['sin_month'] = df['month'].apply(lambda x: np.sin(x / 12 * 2 * np.pi))
df['sin_day'] = df['day'].apply(lambda x: np.sin(x / 31 * 2 * np.pi))
df['sin_hour'] = df['hour'].apply(lambda x: np.sin(x / 24 * 2 * np.pi))
df.head()
```

```python
# Subtract minimum year from year column

df['year'] = df['year'] - df['year'].min()
df.head()
```

```python
# Create a new column y that is the tx_count of the next day

df['y'] = df['tx_count'].shift(-24)
df = df.dropna()
df.tail(25)
```

```python
# Do a train/test split by taking the first 80% of the data as the train dataframe and the last 20% as the test dataframe

from sklearn.model_selection import train_test_split

train, test = train_test_split(df, test_size=0.2, random_state=42, shuffle=False)

display(train.head())
display(test.head())
```

```python
# Write the train and test dataframes to csv files

train.to_csv(TRAIN_SET, index=False)
test.to_csv(TEST_SET, index=False)
```
